## 3d Surface

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
print(df3)
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
import numpy as np
theta=np.arange(0,10,0.5)
X = np.sin(theta)
Y = np.cos(theta)
Z = X**2+Y**2
df3 = pd.DataFrame({'x':X,'y':Y,'z':Z})
print(df3.head())
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
#y=10*X + 3
x=np.arange(0,10)
y=10*x+3
```

```python
b0=np.arange(0,6,0.01)
b1=np.arange(7,13,0.01)
```

```python
b0
```

```python
yhat=b0[0]+b1[0]*x
print(yhat)
np.sum(np.square(y-yhat))
```

```python
sse=[]
B0=[]
B1=[]
for j in np.arange(0,len(b0)):
    for i in np.arange(0,len(b1)):
        yhat=b0[j]+b1[i]*x
        #print(yhat)
        err=np.sum(np.square(y-yhat))
        B0=B0+[b0[j]]
        B1=B1+[b1[i]]
        sse=sse+[err]
```

```python
#np.arange(0,len(b0),2)
```

```python
df3 = pd.DataFrame({'b0':B0,'b1':B1,'sse':sse})
df3.shape
df3.iplot(kind='surface',colorscale='rdylbu')
```
