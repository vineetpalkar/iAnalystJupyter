## i) Read an Image :-

    ==>> Flag - It specifies the way in which image should be read, default value is cv2.IMREAD_COLOR..,

                                      Alternatively, we can pass integer value 1 for this flag..,
            => cv2.IMREAD_GRAYSCALE - Loads image in grayscale mode..,
                                      Alternatively, we can pass integer value 0 for this flag..,
            => cv2.IMREAD_UNCHANGED - Loads image as such including alpha channel..,
                                      Alternatively, we can pass integer value -1 for this flag..,

    => It simply destroys all the windows we created. If you want to destroy any specific window, use the

## Color Spaces in OpenCV | Python
    ==>> Color spaces are a way to represent the color channels present in the image that gives the image that particular hue. There are several different color spaces and each has its own significance..,
### i) BGR Color Space : (Blue, Green, Red)
    ==>> OpenCV’s default color space is RGB. However, it actually stores color in the BGR format.
         It is an additive color model where the different intensities of Blue, Green and Red give different
         shades of color..,
### ii) HSV Color Space : (Hue, Saturation, Value)
    ==>> It stores color information in a cylindrical representation of RGB color points. It attempts to depict the colors as perceived by the human eye.
    ==>> Hue value varies from 0-179.,
    ==>> Saturation value varies from 0-255.,
    ==>> Value value varies from 0-255.,
    ==>> It is mostly used for color segmentation purpose..,
### iii) CMYK Color Space : (Cyan, Magenta, Yellow, Black)
    ==>> Unlike, RGB it is a subtractive color space. The CMYK model works by partially or entirely masking
         colors on a lighter, usually white, background. The ink reduces the light that would otherwise be
         reflected. Such a model is called subtractive because inks “subtract” the colors red, green and blue
         from white light. White light minus red leaves cyan, white light minus green leaves magenta, and
         white light minus blue leaves yellow..,

```python
###  Visualizing the different color channels of an RGB image..,
import numpy as np
import cv2
# Loads a color image as it is..,
image_01 = cv2.imread('TengaVally.jpeg', 1)
B, G, R  = cv2.split(image_01)

#Creating Matrix of all zeros of size B, G & R for merging with individual colors..,
zeros = np.zeros(B.shape, np.uint8)


# Creating B, G & R as individual Channel with making others channels to zero..,
B = cv2.merge((B, zeros, zeros))
G = cv2.merge((zeros, G, zeros))
R = cv2.merge((zeros, zeros, R))

# Display Original Image..,
cv2.namedWindow('Original', cv2.WINDOW_NORMAL)
cv2.imshow('Original', image_01)
#cv2.waitKey(8000)

cv2.namedWindow('Blue', cv2.WINDOW_NORMAL)
cv2.imshow('Blue', B)
#print(B)
#cv2.waitKey(5000)


cv2.namedWindow('Green', cv2.WINDOW_NORMAL)
cv2.imshow('Green', G)
#cv2.waitKey(0)

cv2.namedWindow('Red', cv2.WINDOW_NORMAL)
cv2.imshow('Red', R)

cv2.waitKey(0)
cv2.destroyAllWindows()
# Display '''
```

## i) Capture Video from Camera :-

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == 27: #as code for esc is 27  # also can use ==ord('q'): so that it'll exit if q is pressed
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,1)  #0 means flipping around the x-axis and positive value (for example, 1) means flipping around y-axis. Negative value (for example, -1) means flipping around both axes.

        # write the flipped frame
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```

## Code (Common Arguments) :-
    ==>> img         - Image where we wants to draw shapes.,
    ==>> start_point - It is the starting coordinates of The SHAPE. The coordinates are represented as tuples
                       of two values i.e. (X coordinate value, Y coordinate value)..,
         end_point   - It is the ending coordinates of The SHAPE. The coordinates are represented as tuples of
                       two values i.e. (X coordinate value, Y coordinate value).
         color       - Color of the shape..,
                     > For BGR, pass it as a tuple, eg: (255,0,0) for blue.,
                     > For grayscale, just pass the scalar value.,
         thickness   - Thickness of the shape., Default thickness = 1.
                       If -1 is passed for closed figures like circles, it will fill the shape.,
         lineType    - Type of line, whether 8-connected, anti-aliased line etc. By default - 8-connected.,
                     > cv2.LINE_AA gives anti-aliased line which looks great for curves..,

## i) Drawing Line :-
## cv2.line(image, start_point, end_point, color, thickness)

```python
### Create a black image and draw a blue line on it from top-left to bottom-right corners..,
import numpy as np
import cv2

# Create blank image.,
image = np.zeros((512, 512, 3), np.uint8)  #background is black
#image = np.ones((512, 512, 3), np.uint8)*255  #background is white as matrix of ones multiplied by 255 will give white.

# Draw a diagonal blue line with thickness of 5 px.,
mod_image = cv2.line(image, (0, 0), (400, 200), (0, 255, 0), 2)

# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)  #to convert we split RGB as R G B and then shuffle & merge them as we like.
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```

## ii) Drawing Rectangle :-
## cv2.rectangle(image, start_point, end_point, color, thickness)

```python
# Create blank image.,
image = np.zeros((512, 512, 3), np.uint8)

# draw a green rectangle at the top-right corner of image..,
mod_image = cv2.rectangle(image, (300, 0), (510, 150), (0, 255, 0), -1) #the last value can be any negative value. if 0 is entered it will show only border.
# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```

## iv) Drawing Ellipse :-
## cv2.ellipse(image, centerCoordinates, axesLength, angle, startAngle, endAngle, color [, thickness[, lineType[, shift]]])
    => centerCoordinates - It is the center coordinates of ellipse. The coordinates are represented as tuples
                           of two values i.e. (X coordinate value, Y coordinate value).,
       axesLength        - It contains tuple of two variables containing major and minor axis of ellipse
                           (major axis length, minor axis length).,
       angle             - Ellipse rotation angle in degrees.,
       startAngle        - Starting angle of the elliptic arc in degrees.,
       endAngle          - Ending angle of the elliptic arc in degrees.,
       lineType          - This is an optional parameter. It gives the type of the ellipse boundary.,
       shift             - This is an optional parameter. It denotes the number of fractional bits in the
                           coordinates of the center and values of axes.,

```python
# Create blank image.,
import pandas as pd
image = np.zeros((512, 512, 3), np.uint8)

# draw a half ellipse at the center of the image..,
mod_image = cv2.ellipse(image, (260,200), (250,40), 0, 360, 0, (0, 255, 0), 5)


# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```

## vi) Adding Text to Images :
## cv2.putText(image, text, org, font, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])
    => text - Text string to be drawn.,
       org  - It is the coordinates of the bottom-left corner of the text string in the image. The coordinates
              are represented as tuples of two values i.e. (X coordinate value, Y coordinate value).,
       font - It denotes the font type. Some of font types are FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN etc.,
       fontScale - Font scale factor that is multiplied by the font-specific base size.,

```python
# Create blank image.,
image = np.zeros((512, 512, 3), np.uint8)

# font 
font = cv2.FONT_HERSHEY_SIMPLEX
# org 
org = (50, 250)
# fontScale 
fontScale = 1 
# Blue color in BGR
color = (255, 0, 0)  
# Line thickness of 2 px 
thickness = 5

mod_image = cv2.putText(image, 'HelloOpenCV', org, font, fontScale, color, thickness, cv2.LINE_AA)

# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```

## Ex. Draw a triangle with centroid using OpenCV:-

```python
import cv2
import numpy as np

# Width and height of the black window  
width = 500
height = 500

# Create Black window..,
image = np.zeros((height, width, 3), np.uint8)

# Three vertices of Triangle..,
P1 = (200, 100)
P2 = (100, 100)
P3 = (250, 400)

# Drawing three lines with three vertices to form triangle..,
cv2.line(image, P1, P2, (255, 0, 0), 3)
cv2.line(image, P2, P3, (0, 255, 0), 3)
cv2.line(image, P3, P1, (0, 0, 255), 3)

# Finding Centroid of the triangle..,
Centroid = (int((P1[0] + P2[0] + P3[0])/3), int((P1[1] + P2[1] + P3[1])/3))
print('Coordinates of Centroid : ', Centroid)

# Draw Centroid..,
cv2.circle(image, Centroid, 3, (255, 255, 255), 3)

# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(image)
image = cv2.merge([R, G, B])
plt.imshow(image)
```
