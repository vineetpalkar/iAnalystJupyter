```python
### Create a black image and draw a blue line on it from top-left to bottom-right corners..,
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Create blank image.,
image = np.ones((512, 512, 3), np.uint8)*255


# Draw a diagonal blue line with thickness of 5 px.,
mod_image = cv2.line(image, (0, 0), (400, 200), (0, 0, 255), 5)


# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```

```python
# Create blank image.,
image = np.zeros((512, 512, 3), np.uint8)

# draw a green rectangle at the top-right corner of image..,
mod_image = cv2.rectangle(image, (300, 0), (500, 150), (0, 255, 0), -255)  #if the last value of -255 is kept 0 then it'll show only border. try keeping 255

# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```

```python
# draw a circle inside the rectangle drawn above.,
mod_image = cv2.circle(mod_image, (405, 75), 65, (0, 0, 255), -1)

# Convert BGR (OpenCV) to RGB (MatPlotLib).,
B, G, R = cv2.split(mod_image)
mod_image = cv2.merge([R, G, B])
plt.imshow(mod_image)
```
