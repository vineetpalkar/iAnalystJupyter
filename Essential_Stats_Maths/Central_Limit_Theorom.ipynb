```python
import numpy as np
import pandas as pd
```

```python
x=pd.Series(np.random.rand(200))
```

```python
x.plot(kind='kde')
```

```python
ss=10
idx=np.random.randint(0,200,ss)
idx 
```

```python
x[idx]   #taking out 10 samples out of popln using their corresponding index
```

```python
x[idx].mean()
```

```python
ns=500  #no. of samples  #change no. of samples
ss= 10  # sample size #change sample sizes. Increasing the sample size will result in distn being bell shaped curve and not no. of samples
m=[]
for i in range(ns):
  idx=np.random.randint(0,200,ss)
  m.append(x[idx].mean())
m=pd.Series(m)
m.plot(kind= 'kde') #try kind='hist'
```
